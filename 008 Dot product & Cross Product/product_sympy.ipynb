# 點積與叉積 (Dot Product & Cross Product)

本筆記本使用 **SymPy** 介紹向量的兩種核心乘法：

1. **點積（內積）定義** — $\mathbf{u}\cdot\mathbf{v}$
2. **幾何意義** — 夾角、正交、投影
3. **點積性質** — 交換、分配、與範數
4. **叉積定義** — $\mathbf{u}\times\mathbf{v}$（僅 $\mathbb{R}^3$）
5. **幾何意義** — 法向量、平行四邊形面積、右手定則
6. **叉積性質** — 反交換、分配、與點積的關係
7. **純量三重積** — $(\mathbf{u}\times\mathbf{v})\cdot\mathbf{w}$ 與平行六面體體積
8. **應用** — 平面法向量、面積、體積
9. **綜合練習**

> SymPy 做的是**符號／精確**計算。需要大量數值陣列運算時請看同目錄的 `product_numpy.ipynb`。


```python
import sympy as sp
from IPython.display import display, Math


def display_large(expr):
    """Display a sympy expression in LaTeX with large font size."""
    display(Math(r"\large{%s}" % sp.latex(expr)))


def display_huge(expr):
    """Display a sympy expression in LaTeX with huge font size."""
    display(Math(r"\huge{%s}" % sp.latex(expr)))
```

## 1. 點積（內積）定義

在 $\mathbb{R}^n$ 中，兩向量

$$
\mathbf{u}=\begin{bmatrix}u_1\\\vdots\\u_n\end{bmatrix},\quad
\mathbf{v}=\begin{bmatrix}v_1\\\vdots\\v_n\end{bmatrix}
$$

的**點積（dot product / inner product）**為一個**純量**：

$$
\mathbf{u}\cdot\mathbf{v}
= u_1 v_1 + u_2 v_2 + \cdots + u_n v_n
= \mathbf{u}^{\mathsf{T}}\mathbf{v}
$$

| 名稱 | 英文 | 結果 |
|------|------|------|
| **點積／內積** | dot / inner product | 純量（scalar） |

SymPy：`u.dot(v)` 或 `u.T * v`。


```python
u = sp.Matrix([1, 2, 3])
v = sp.Matrix([4, -5, 6])

print("u =")
display_huge(u)
print("v =")
display_huge(v)

print("u · v =")
display_huge(u.dot(v))

print("via u^T v =")
display_huge((u.T * v)[0])
```

## 2. 幾何意義：夾角、正交、投影

點積與夾角 $\theta$ 的關係：

$$
\mathbf{u}\cdot\mathbf{v} = \|\mathbf{u}\|\|\mathbf{v}\|\cos\theta
\quad\Rightarrow\quad
\cos\theta = \frac{\mathbf{u}\cdot\mathbf{v}}{\|\mathbf{u}\|\|\mathbf{v}\|}
$$

因此：

| 條件 | 意義 |
|------|------|
| $\mathbf{u}\cdot\mathbf{v}=0$ | **正交**（垂直），$\theta=90^\circ$ |
| $\mathbf{u}\cdot\mathbf{v}>0$ | 銳角 |
| $\mathbf{u}\cdot\mathbf{v}<0$ | 鈍角 |

**範數**由點積定義：

$$
\|\mathbf{u}\| = \sqrt{\mathbf{u}\cdot\mathbf{u}}
$$

**投影**（$\mathbf{v}$ 投影到 $\mathbf{u}$）：

$$
\mathrm{proj}_{\mathbf{u}}\mathbf{v}
= \frac{\mathbf{u}\cdot\mathbf{v}}{\|\mathbf{u}\|^2}\,\mathbf{u}
$$


```python
u = sp.Matrix([3, 0])
v = sp.Matrix([2, 2])

dot = u.dot(v)
nu = u.norm()
nv = v.norm()
cos_theta = sp.simplify(dot / (nu * nv))
theta = sp.acos(cos_theta)

print("u =")
display_huge(u)
print("v =")
display_huge(v)
print("u · v =")
display_huge(dot)
print("||u||, ||v|| =")
display_huge(sp.Matrix([nu, nv]))
print("cos θ =")
display_huge(cos_theta)
print("θ (radians) =")
display_huge(theta)
print("θ (degrees) =")
display_huge(sp.deg(theta))

# Orthogonal check
a = sp.Matrix([1, 2, -1])
b = sp.Matrix([2, -1, 0])
print("\nOrthogonal example: a · b =")
display_huge(a.dot(b))
print("Orthogonal?", a.dot(b) == 0)

# Projection
proj = (u.dot(v) / u.dot(u)) * u
print("\nproj_u(v) =")
display_huge(proj)
print("orthogonal remainder v - proj =")
display_huge(sp.simplify(v - proj))
print("remainder · u =")
display_huge(sp.simplify((v - proj).dot(u)))
```

## 3. 點積的性質

對任意向量 $\mathbf{u},\mathbf{v},\mathbf{w}$ 與純量 $c$：

| 性質 | 公式 |
|------|------|
| 交換律 | $\mathbf{u}\cdot\mathbf{v}=\mathbf{v}\cdot\mathbf{u}$ |
| 分配律 | $\mathbf{u}\cdot(\mathbf{v}+\mathbf{w})=\mathbf{u}\cdot\mathbf{v}+\mathbf{u}\cdot\mathbf{w}$ |
| 純量 | $(c\mathbf{u})\cdot\mathbf{v}=c(\mathbf{u}\cdot\mathbf{v})$ |
| 正定 | $\mathbf{u}\cdot\mathbf{u}=\|\mathbf{u}\|^2\geq 0$，等號 $\Leftrightarrow\mathbf{u}=\mathbf{0}$ |
| Cauchy–Schwarz | $|\mathbf{u}\cdot\mathbf{v}|\leq\|\mathbf{u}\|\|\mathbf{v}\|$ |

注意：點積**沒有**「結合律」這種三個向量連乘成一個向量的說法——結果永遠是純量。


```python
u = sp.Matrix(sp.symbols("u1:4"))
v = sp.Matrix(sp.symbols("v1:4"))
w = sp.Matrix(sp.symbols("w1:4"))
c = sp.symbols("c")

print("Commutative: u·v - v·u =")
display_large(sp.simplify(u.dot(v) - v.dot(u)))

print("Distributive: u·(v+w) - (u·v + u·w) =")
display_large(sp.simplify(u.dot(v + w) - (u.dot(v) + u.dot(w))))

print("Scalar: (c u)·v - c(u·v) =")
display_large(sp.simplify((c * u).dot(v) - c * u.dot(v)))

# Numerical Cauchy–Schwarz check
p = sp.Matrix([1, -2, 3])
q = sp.Matrix([4, 0, -1])
print("\n|Cauchy–Schwarz|: |p·q| =", abs(p.dot(q)), ", ||p||||q|| =", p.norm() * q.norm())
print("Holds?", abs(p.dot(q)) <= p.norm() * q.norm())
```

## 4. 叉積定義（僅 $\mathbb{R}^3$）

兩向量 $\mathbf{u},\mathbf{v}\in\mathbb{R}^3$ 的**叉積（cross product）**是另一個**向量**：

$$
\mathbf{u}\times\mathbf{v}
=
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3
\end{vmatrix}
=
\begin{bmatrix}
u_2 v_3 - u_3 v_2 \\
u_3 v_1 - u_1 v_3 \\
u_1 v_2 - u_2 v_1
\end{bmatrix}
$$

| 名稱 | 英文 | 結果 | 空間 |
|------|------|------|------|
| **叉積** | cross product | 向量 | 主要在 $\mathbb{R}^3$ |

SymPy：`u.cross(v)`。


```python
u = sp.Matrix([1, 2, 3])
v = sp.Matrix([4, 5, 6])

print("u =")
display_huge(u)
print("v =")
display_huge(v)
print("u × v =")
display_huge(u.cross(v))

# Manual formula check
manual = sp.Matrix([
    u[1] * v[2] - u[2] * v[1],
    u[2] * v[0] - u[0] * v[2],
    u[0] * v[1] - u[1] * v[0],
])
print("manual formula =")
display_huge(manual)
print("Match?", u.cross(v) == manual)
```

## 5. 叉積的幾何意義

$$
\|\mathbf{u}\times\mathbf{v}\| = \|\mathbf{u}\|\|\mathbf{v}\|\sin\theta
$$

等於以 $\mathbf{u},\mathbf{v}$ 為鄰邊的**平行四邊形面積**。

方向由**右手定則**決定：$\mathbf{u}\times\mathbf{v}$ **垂直**於 $\mathbf{u}$ 與 $\mathbf{v}$ 張成的平面，

$$
(\mathbf{u}\times\mathbf{v})\cdot\mathbf{u}=0,\qquad
(\mathbf{u}\times\mathbf{v})\cdot\mathbf{v}=0
$$

特殊情況：

- $\mathbf{u}\parallel\mathbf{v}$ $\Rightarrow$ $\mathbf{u}\times\mathbf{v}=\mathbf{0}$（$\sin\theta=0$）
- $\mathbf{u}\perp\mathbf{v}$ $\Rightarrow$ $\|\mathbf{u}\times\mathbf{v}\|=\|\mathbf{u}\|\|\mathbf{v}\|$


```python
u = sp.Matrix([1, 0, 0])
v = sp.Matrix([1, 1, 0])
n = u.cross(v)

print("u =")
display_huge(u)
print("v =")
display_huge(v)
print("n = u × v =")
display_huge(n)

print("n · u =")
display_huge(n.dot(u))
print("n · v =")
display_huge(n.dot(v))

area = n.norm()
print("parallelogram area ||u × v|| =")
display_huge(area)
sin_theta = sp.sin(sp.acos(u.dot(v) / (u.norm() * v.norm())))
print("||u||||v||sinθ =")
display_huge(sp.simplify(u.norm() * v.norm() * sin_theta))

# Parallel => zero cross
a = sp.Matrix([1, 2, 3])
b = 5 * a
print("\nParallel: a × (5a) =")
display_huge(a.cross(b))
```

## 6. 叉積的性質

| 性質 | 公式 |
|------|------|
| 反交換 | $\mathbf{u}\times\mathbf{v}=-\mathbf{v}\times\mathbf{u}$ |
| 分配律 | $\mathbf{u}\times(\mathbf{v}+\mathbf{w})=\mathbf{u}\times\mathbf{v}+\mathbf{u}\times\mathbf{w}$ |
| 純量 | $(c\mathbf{u})\times\mathbf{v}=c(\mathbf{u}\times\mathbf{v})$ |
| 自叉 | $\mathbf{u}\times\mathbf{u}=\mathbf{0}$ |
| 不結合 | $(\mathbf{u}\times\mathbf{v})\times\mathbf{w}\neq\mathbf{u}\times(\mathbf{v}\times\mathbf{w})$（一般） |

與點積的重要恆等式（Lagrange）：

$$
\|\mathbf{u}\times\mathbf{v}\|^2
= \|\mathbf{u}\|^2\|\mathbf{v}\|^2-(\mathbf{u}\cdot\mathbf{v})^2
$$


```python
u = sp.Matrix(sp.symbols("u1:4"))
v = sp.Matrix(sp.symbols("v1:4"))
w = sp.Matrix(sp.symbols("w1:4"))
c = sp.symbols("c")

print("Anti-commutative: u×v + v×u =")
display_large(sp.simplify(u.cross(v) + v.cross(u)))

print("Distributive remainder:")
display_large(sp.simplify(u.cross(v + w) - (u.cross(v) + u.cross(w))))

print("Self cross u×u =")
display_large(u.cross(u))

# Not associative (numeric counterexample)
a = sp.Matrix([1, 0, 0])
b = sp.Matrix([1, 1, 0])
cvec = sp.Matrix([0, 1, 1])
left = a.cross(b).cross(cvec)
right = a.cross(b.cross(cvec))
print("\n(a×b)×c =")
display_huge(left)
print("a×(b×c) =")
display_huge(right)
print("Equal?", left == right)

# Lagrange identity
p = sp.Matrix([2, -1, 3])
q = sp.Matrix([1, 4, -2])
lhs = p.cross(q).dot(p.cross(q))
rhs = p.dot(p) * q.dot(q) - p.dot(q) ** 2
print("\nLagrange: ||p×q||² =", lhs, ", ||p||²||q||²-(p·q)² =", rhs)
print("Match?", lhs == rhs)
```

## 7. 純量三重積

$$
(\mathbf{u}\times\mathbf{v})\cdot\mathbf{w}
=
\det\begin{bmatrix}
\mathbf{u} & \mathbf{v} & \mathbf{w}
\end{bmatrix}
=
\begin{vmatrix}
u_1 & v_1 & w_1 \\
u_2 & v_2 & w_2 \\
u_3 & v_3 & w_3
\end{vmatrix}
$$

幾何意義：以三向量為鄰邊的**平行六面體有號體積**。

- $=0$ $\Leftrightarrow$ 三向量共面（線性相依）
- 絕對值 = 體積


```python
u = sp.Matrix([1, 0, 0])
v = sp.Matrix([1, 2, 0])
w = sp.Matrix([1, 2, 3])

triple = u.cross(v).dot(w)
M = sp.Matrix.hstack(u, v, w)

print("u, v, w as columns of M:")
display_huge(M)
print("(u × v) · w =")
display_huge(triple)
print("det(M) =")
display_huge(M.det())
print("Match?", triple == M.det())

print("volume = |triple| =")
display_huge(sp.Abs(triple))

# Coplanar => zero
w_coplanar = u + 2 * v
print("\nCoplanar w = u+2v, triple =")
display_huge(u.cross(v).dot(w_coplanar))
```

## 8. 應用：平面法向量、面積、體積

1. **平面法向量**：平面上兩方向 $\mathbf{u},\mathbf{v}$，則 $\mathbf{n}=\mathbf{u}\times\mathbf{v}$ 垂直該平面。
2. **三角形面積**：$\dfrac{1}{2}\|\mathbf{u}\times\mathbf{v}\|$
3. **平行六面體體積**：$|(\mathbf{u}\times\mathbf{v})\cdot\mathbf{w}|$


```python
# Plane through P0 with directions u, v
P0 = sp.Matrix([1, 1, 1])
u = sp.Matrix([1, 0, -1])
v = sp.Matrix([0, 1, 2])
n = u.cross(v)

print("normal n = u × v =")
display_huge(n)

# Plane equation: n · (X - P0) = 0
x, y, z = sp.symbols("x y z")
X = sp.Matrix([x, y, z])
plane = sp.expand(n.dot(X - P0))
print("plane equation n·(X-P0) = 0:")
display_huge(sp.Eq(plane, 0))

# Triangle area
area = sp.Rational(1, 2) * n.norm()
print("triangle area =")
display_huge(area)

# Parallelepiped volume
w = sp.Matrix([2, -1, 1])
vol = sp.Abs(u.cross(v).dot(w))
print("parallelepiped volume =")
display_huge(vol)
```

## 9. 綜合練習

給定

$$
\mathbf{a}=\begin{bmatrix}1\\2\\-1\end{bmatrix},\quad
\mathbf{b}=\begin{bmatrix}0\\1\\3\end{bmatrix},\quad
\mathbf{c}=\begin{bmatrix}2\\-1\\1\end{bmatrix}
$$

1. 計算 $\mathbf{a}\cdot\mathbf{b}$，並求夾角 $\theta$（度）
2. 計算 $\mathbf{a}\times\mathbf{b}$，驗證它同時垂直於 $\mathbf{a}$ 與 $\mathbf{b}$
3. 求以 $\mathbf{a},\mathbf{b}$ 為鄰邊的平行四邊形面積
4. 計算純量三重積 $(\mathbf{a}\times\mathbf{b})\cdot\mathbf{c}$，並與 $\det[\mathbf{a}\ \mathbf{b}\ \mathbf{c}]$ 比對
5. 求 $\mathrm{proj}_{\mathbf{a}}\mathbf{b}$


```python
a = sp.Matrix([1, 2, -1])
b = sp.Matrix([0, 1, 3])
c = sp.Matrix([2, -1, 1])

# 1. Dot and angle
dot = a.dot(b)
cos_theta = sp.simplify(dot / (a.norm() * b.norm()))
theta_deg = sp.deg(sp.acos(cos_theta))
print("1. a · b =")
display_huge(dot)
print("θ (degrees) =")
display_huge(theta_deg)

# 2. Cross and orthogonality
cross = a.cross(b)
print("\n2. a × b =")
display_huge(cross)
print("(a×b)·a =", cross.dot(a), ", (a×b)·b =", cross.dot(b))

# 3. Area
print("\n3. parallelogram area =")
display_huge(cross.norm())

# 4. Triple product
triple = cross.dot(c)
M = sp.Matrix.hstack(a, b, c)
print("\n4. (a×b)·c =")
display_huge(triple)
print("det([a b c]) =")
display_huge(M.det())
print("Match?", triple == M.det())

# 5. Projection
proj = (a.dot(b) / a.dot(a)) * a
print("\n5. proj_a(b) =")
display_huge(proj)
```

## 小結

| 概念 | SymPy / 公式 | 重點 |
|------|--------------|------|
| 點積 | `u.dot(v)` / $\mathbf{u}^{\mathsf{T}}\mathbf{v}$ | 結果是**純量** |
| 夾角 | $\cos\theta=\frac{\mathbf{u}\cdot\mathbf{v}}{\|\mathbf{u}\|\|\mathbf{v}\|}$ | $=0$ ⇒ 正交 |
| 投影 | $\mathrm{proj}_{\mathbf{u}}\mathbf{v}=\frac{\mathbf{u}\cdot\mathbf{v}}{\|\mathbf{u}\|^2}\mathbf{u}$ | 餘量垂直於 $\mathbf{u}$ |
| 叉積 | `u.cross(v)` | 結果是**向量**（$\mathbb{R}^3$） |
| 面積 | $\|\mathbf{u}\times\mathbf{v}\|$ | 平行四邊形面積 |
| 法向 | $\mathbf{u}\times\mathbf{v}\perp\mathbf{u},\mathbf{v}$ | 右手定則 |
| 三重積 | `(u.cross(v)).dot(w)` / `det` | 平行六面體有號體積 |

**記住：**

- 點積：量「有多平行」；叉積：量「有多垂直」並給出法向
- $\mathbf{u}\cdot\mathbf{v}=0$ ⇒ 正交；$\mathbf{u}\times\mathbf{v}=\mathbf{0}$ ⇒ 平行
- 叉積反交換：$\mathbf{u}\times\mathbf{v}=-\mathbf{v}\times\mathbf{u}$
- 體積用 $|(\mathbf{u}\times\mathbf{v})\cdot\mathbf{w}|$
