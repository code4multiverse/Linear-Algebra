# 正交 (Orthogonal)

本筆記本延續線性獨立，專門討論**正交**：

1. **內積與正交定義** — 何時兩向量「垂直」
2. **正交集合／正規正交集合** — Orthogonal vs Orthonormal
3. **與線性獨立的關係** — 非零正交向量一定獨立
4. **正交投影** — 把向量分解成平行＋垂直分量
5. **四大子空間的正交性** — 列空間 $\perp$ 左零空間等
6. **正交矩陣** — $Q^{\mathsf{T}}Q=I$
7. **Gram–Schmidt** — 從任意基底造出正交基底
8. **綜合練習**

```python
import sympy as sp
from IPython.display import display, Math


def display_large(expr):
    """Display a sympy expression in LaTeX with large font size."""
    display(Math(r"\large{%s}" % sp.latex(expr)))


def display_huge(expr):
    """Display a sympy expression in LaTeX with huge font size."""
    display(Math(r"\huge{%s}" % sp.latex(expr)))
```

## 1. 內積複習

在 $\mathbb{R}^n$ 中，兩向量 $\mathbf{u},\mathbf{v}$ 的**內積（dot product）**為

$$
\mathbf{u}\cdot\mathbf{v}
= \mathbf{u}^{\mathsf{T}}\mathbf{v}
= u_1 v_1 + u_2 v_2 + \cdots + u_n v_n
$$

長度（歐氏範數）為

$$
\|\mathbf{u}\| = \sqrt{\mathbf{u}\cdot\mathbf{u}}
$$

幾何關係：

$$
\mathbf{u}\cdot\mathbf{v} = \|\mathbf{u}\|\|\mathbf{v}\|\cos\theta
$$

其中 $\theta$ 是兩向量夾角。

```python
u = sp.Matrix([3, 4])
v = sp.Matrix([4, -3])

print("u =")
display_huge(u)
print("v =")
display_huge(v)

dot = u.dot(v)
print("u · v =")
display_large(dot)

print("||u|| =")
display_large(u.norm())
print("||v|| =")
display_large(v.norm())
```

## 2. 什麼是正交？

| 名稱 | 英文 | 定義 |
|------|------|------|
| **正交** | Orthogonal | $\mathbf{u}\cdot\mathbf{v} = 0$（夾角 $90^\circ$） |
| **正規正交** | Orthonormal | 兩兩正交，且每個長度為 $1$ |

注意：

- $\mathbf{0}$ 與**任意**向量正交（因為 $\mathbf{0}\cdot\mathbf{v}=0$）
- 「正交」比「不平行」更強：不平行只是方向不同；正交是夾角剛好 $90^\circ$

```python
def are_orthogonal(a, b):
    """Return True if a · b = 0."""
    return a.dot(b) == 0


a = sp.Matrix([1, 2])
b = sp.Matrix([2, -1])  # orthogonal to a
c = sp.Matrix([3, 1])   # not orthogonal to a

print("a · b =")
display_large(a.dot(b))
print("orthogonal(a, b)?", are_orthogonal(a, b))

print("a · c =")
display_large(a.dot(c))
print("orthogonal(a, c)?", are_orthogonal(a, c))
```

## 3. 正交集合與正規正交集合

一組向量 $\{\mathbf{v}_1,\ldots,\mathbf{v}_k\}$ 稱為

- **正交集合**：對所有 $i\neq j$，$\mathbf{v}_i\cdot\mathbf{v}_j = 0$
- **正規正交集合**：正交，且 $\|\mathbf{v}_i\|=1$ 對所有 $i$

標準基底 $\{\mathbf{e}_1,\ldots,\mathbf{e}_n\}$ 就是 $\mathbb{R}^n$ 最常見的正規正交基底。

若把正規正交向量排成矩陣 $Q$ 的行，則

$$
Q^{\mathsf{T}}Q = I
$$

（行與行的內積組成單位矩陣）。

```python
def check_orthogonal_set(vectors, names=None):
    """Check pairwise orthogonality and whether the set is orthonormal."""
    k = len(vectors)
    if names is None:
        names = [f"v{i+1}" for i in range(k)]

    A = sp.Matrix.hstack(*vectors)
    print("Matrix with columns", ", ".join(names) + ":")
    display_huge(A)

    print("Gram matrix G = A^T A (entries = pairwise dots):")
    G = sp.simplify(A.T * A)
    display_huge(G)

    orthogonal = all(G[i, j] == 0 for i in range(k) for j in range(k) if i != j)
    orthonormal = orthogonal and all(G[i, i] == 1 for i in range(k))

    print("Orthogonal set?", orthogonal)
    print("Orthonormal set?", orthonormal)
    print()
    return A, G


# Orthogonal but not yet unit length
v1 = sp.Matrix([1, 1, 0])
v2 = sp.Matrix([1, -1, 0])
v3 = sp.Matrix([0, 0, 1])

check_orthogonal_set([v1, v2, v3], names=["v1", "v2", "v3"])

# Normalize to get orthonormal set
u1 = v1 / v1.norm()
u2 = v2 / v2.norm()
u3 = v3 / v3.norm()

print("After normalizing:")
Q, GQ = check_orthogonal_set([u1, u2, u3], names=["u1", "u2", "u3"])
```

## 4. 正交 $\Rightarrow$ 線性獨立

**定理：** 若 $\{\mathbf{v}_1,\ldots,\mathbf{v}_k\}$ 兩兩正交，且每個 $\mathbf{v}_i\neq\mathbf{0}$，則它們**線性獨立**。

證明想法：假設 $c_1\mathbf{v}_1+\cdots+c_k\mathbf{v}_k=\mathbf{0}$。
兩邊與 $\mathbf{v}_j$ 取內積：

$$
c_j\|\mathbf{v}_j\|^2 = 0
\quad\Rightarrow\quad
c_j = 0
$$

對每個 $j$ 成立，故只有平凡解。

| 關係 | 說明 |
|------|------|
| 正交（非零） | $\Rightarrow$ 線性獨立 |
| 線性獨立 | **不保證**正交（例如 $(1,0)$ 與 $(1,1)$） |

正交是「特別好用」的獨立集合：係數可用內積直接讀出。

```python
# Orthogonal nonzero => independent
A_orth = sp.Matrix.hstack(v1, v2, v3)
print("Orthogonal columns:")
display_huge(A_orth)
print("rank =", A_orth.rank(), ", cols =", A_orth.cols)
print("Independent?", A_orth.rank() == A_orth.cols)
print("Nullspace:", A_orth.nullspace())

# Independent but NOT orthogonal
w1 = sp.Matrix([1, 0])
w2 = sp.Matrix([1, 1])
A_ind = sp.Matrix.hstack(w1, w2)
print("\nIndependent but not orthogonal:")
display_huge(A_ind)
print("rank =", A_ind.rank())
print("w1 · w2 =")
display_large(w1.dot(w2))
```

## 5. 正交展開：係數超好算

若 $\{\mathbf{q}_1,\ldots,\mathbf{q}_k\}$ 正規正交，且

$$
\mathbf{b} = c_1\mathbf{q}_1 + \cdots + c_k\mathbf{q}_k
$$

則

$$
c_j = \mathbf{q}_j\cdot\mathbf{b}
$$

若只是正交（尚未單位化）：

$$
c_j = \frac{\mathbf{v}_j\cdot\mathbf{b}}{\|\mathbf{v}_j\|^2}
$$

不必解線性方程組——這是正交基底的核心優勢。

```python
b = sp.Matrix([2, 4, 6])

# Expand b in orthonormal basis {u1, u2, u3}
c1 = u1.dot(b)
c2 = u2.dot(b)
c3 = u3.dot(b)

print("b =")
display_huge(b)
print("Coefficients via dots:")
display_large(sp.Matrix([c1, c2, c3]))

recon = c1*u1 + c2*u2 + c3*u3
print("Reconstruction c1*u1 + c2*u2 + c3*u3 =")
display_huge(sp.simplify(recon))
print("Match?", sp.simplify(recon - b) == sp.Matrix([0, 0, 0]))
```

## 6. 正交投影

把 $\mathbf{b}$ 投影到由非零向量 $\mathbf{a}$ 張成的直線：

$$
\operatorname{proj}_{\mathbf{a}}\mathbf{b}
= \frac{\mathbf{a}\cdot\mathbf{b}}{\|\mathbf{a}\|^2}\,\mathbf{a}
$$

誤差向量

$$
\mathbf{e} = \mathbf{b} - \operatorname{proj}_{\mathbf{a}}\mathbf{b}
$$

滿足 $\mathbf{e}\perp\mathbf{a}$。

幾何圖像：任意 $\mathbf{b}$ 可唯一分解成「沿著 $\mathbf{a}$」＋「垂直於 $\mathbf{a}$」。

```python
def proj(a, b):
    """Orthogonal projection of b onto the line spanned by a."""
    return (a.dot(b) / a.dot(a)) * a


a = sp.Matrix([2, 1])
b = sp.Matrix([3, 4])

p = proj(a, b)
e = b - p

print("a =")
display_huge(a)
print("b =")
display_huge(b)
print("proj_a(b) =")
display_huge(sp.simplify(p))
print("error e = b - proj =")
display_huge(sp.simplify(e))
print("e · a =")
display_large(sp.simplify(e.dot(a)))
print("Check: p + e == b?", sp.simplify(p + e - b) == sp.Matrix([0, 0]))
```

### 投影到子空間

若 $A$ 的行線性獨立，投影 $\mathbf{b}$ 到 $\operatorname{Col}(A)$ 的公式為

$$
\hat{\mathbf{x}} = (A^{\mathsf{T}}A)^{-1}A^{\mathsf{T}}\mathbf{b},\qquad
\mathbf{p} = A\hat{\mathbf{x}} = P\mathbf{b}
$$

其中投影矩陣 $P = A(A^{\mathsf{T}}A)^{-1}A^{\mathsf{T}}$。

誤差 $\mathbf{e}=\mathbf{b}-\mathbf{p}$ 正交於 $\operatorname{Col}(A)$（亦即 $A^{\mathsf{T}}\mathbf{e}=\mathbf{0}$）。

```python
A = sp.Matrix([[1, 0],
               [1, 1],
               [1, 2]])
b = sp.Matrix([6, 0, 0])

xhat = (A.T * A).inv() * A.T * b
p = A * xhat
e = b - p
P = A * (A.T * A).inv() * A.T

print("A =")
display_huge(A)
print("b =")
display_huge(b)
print("x-hat =")
display_large(sp.simplify(xhat))
print("projection p =")
display_huge(sp.simplify(p))
print("error e =")
display_huge(sp.simplify(e))
print("A^T e (should be 0) =")
display_large(sp.simplify(A.T * e))
print("Projection matrix P =")
display_huge(sp.simplify(P))
```

## 7. 四大子空間的正交性

對任意 $m\times n$ 矩陣 $A$：

| 子空間 | 所在空間 | 正交關係 |
|--------|----------|----------|
| 列空間 $\operatorname{Col}(A)$ | $\mathbb{R}^m$ | $\operatorname{Col}(A)\perp N(A^{\mathsf{T}})$ |
| 左零空間 $N(A^{\mathsf{T}})$ | $\mathbb{R}^m$ | （同上） |
| 行空間 $\operatorname{Row}(A)$ | $\mathbb{R}^n$ | $\operatorname{Row}(A)\perp N(A)$ |
| 零空間 $N(A)$ | $\mathbb{R}^n$ | （同上） |

直觀：

- $A\mathbf{x}=\mathbf{0}$ $\Leftrightarrow$ $\mathbf{x}$ 與 $A$ 的每一**列（row）**正交
- $A^{\mathsf{T}}\mathbf{y}=\mathbf{0}$ $\Leftrightarrow$ $\mathbf{y}$ 與 $A$ 的每一**行（column）**正交

```python
A = sp.Matrix([[1, 2, 3],
               [2, 4, 6],
               [1, 1, 1]])

print("A =")
display_huge(A)

null_A = A.nullspace()
row_space = [A.row(i).T for i in range(A.rows)]  # as column vectors

print("Nullspace basis of A:")
for i, nvec in enumerate(null_A):
    print(f"  n{i+1} =")
    display_large(nvec)

print("Check: each null vector is orthogonal to every row of A")
for nvec in null_A:
    dots = [sp.simplify(A.row(i).dot(nvec)) for i in range(A.rows)]
    print("  row · n =", dots)

left_null = A.T.nullspace()
print("\nLeft-nullspace basis (nullspace of A^T):")
for i, y in enumerate(left_null):
    print(f"  y{i+1} =")
    display_large(y)

print("Check: each left-null vector is orthogonal to every column of A")
for y in left_null:
    dots = [sp.simplify(A.col(j).dot(y)) for j in range(A.cols)]
    print("  col · y =", dots)
```

## 8. 正交矩陣

方陣 $Q$ 稱為**正交矩陣（orthogonal matrix）**若

$$
Q^{\mathsf{T}}Q = I
\quad\Leftrightarrow\quad
Q^{-1} = Q^{\mathsf{T}}
$$

等價條件：

- $Q$ 的行（columns）構成正規正交基底
- $Q$ 的列（rows）也構成正規正交基底
- $Q$ 保持長度：$\|Q\mathbf{x}\|=\|\mathbf{x}\|$（旋轉／反射）

常見例子：二維旋轉矩陣

$$
R_\theta =
\begin{bmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{bmatrix}
$$


```python
theta = sp.pi / 3
R = sp.Matrix([[sp.cos(theta), -sp.sin(theta)],
               [sp.sin(theta),  sp.cos(theta)]])

print("Rotation matrix R =")
display_huge(sp.simplify(R))

print("R^T R =")
display_huge(sp.simplify(R.T * R))

print("R^{-1} =")
display_huge(sp.simplify(R.inv()))
print("Equals R^T?", sp.simplify(R.inv() - R.T) == sp.zeros(2))

x = sp.Matrix([1, 0])
print("||x|| =")
display_large(x.norm())
print("||R x|| =")
display_large(sp.simplify((R * x).norm()))
```

## 9. Gram–Schmidt 正交化

從線性獨立向量 $\{\mathbf{a}_1,\ldots,\mathbf{a}_k\}$ 造出正交向量 $\{\mathbf{v}_1,\ldots,\mathbf{v}_k\}$：

$$
\begin{aligned}
\mathbf{v}_1 &= \mathbf{a}_1 \\
\mathbf{v}_2 &= \mathbf{a}_2 - \operatorname{proj}_{\mathbf{v}_1}\mathbf{a}_2 \\
\mathbf{v}_3 &= \mathbf{a}_3 - \operatorname{proj}_{\mathbf{v}_1}\mathbf{a}_3 - \operatorname{proj}_{\mathbf{v}_2}\mathbf{a}_3 \\
&\vdots
\end{aligned}
$$

再單位化可得正規正交基底。SymPy：`Matrix.QRdecomposition()` 或逐步手算。

```python
def gram_schmidt(vectors):
    """Classical Gram-Schmidt: return orthogonal vectors spanning the same space."""
    orth = []
    for a in vectors:
        v = sp.Matrix(a)
        for u in orth:
            v = v - proj(u, a)
        v = sp.simplify(v)
        if v != sp.zeros(*v.shape):
            orth.append(v)
    return orth


a1 = sp.Matrix([1, 1, 0])
a2 = sp.Matrix([1, 0, 1])
a3 = sp.Matrix([0, 1, 1])

print("Original (independent) vectors:")
for i, a in enumerate([a1, a2, a3], 1):
    print(f"a{i} =")
    display_large(a)

orth = gram_schmidt([a1, a2, a3])
print("After Gram-Schmidt (orthogonal):")
for i, v in enumerate(orth, 1):
    print(f"v{i} =")
    display_large(v)

check_orthogonal_set(orth, names=[f"v{i}" for i in range(1, len(orth) + 1)])

# Compare with QR
A = sp.Matrix.hstack(a1, a2, a3)
Q, R = A.QRdecomposition()
print("Q from QR (orthonormal columns):")
display_huge(sp.simplify(Q))
print("R =")
display_huge(sp.simplify(R))
print("Q^T Q =")
display_huge(sp.simplify(Q.T * Q))
```

## 10. 綜合練習

考慮

$$
\mathbf{u}_1 = \begin{pmatrix} 1 \\ 2 \\ 2 \end{pmatrix},\quad
\mathbf{u}_2 = \begin{pmatrix} 2 \\ -1 \\ 0 \end{pmatrix},\quad
\mathbf{b} = \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}
$$

請完成：

1. 驗證 $\mathbf{u}_1\perp\mathbf{u}_2$
2. 將 $\{\mathbf{u}_1,\mathbf{u}_2\}$ 單位化成正規正交集合 $\{\mathbf{q}_1,\mathbf{q}_2\}$
3. 計算 $\mathbf{b}$ 在 $\operatorname{span}\{\mathbf{u}_1,\mathbf{u}_2\}$ 上的正交投影 $\mathbf{p}$
4. 驗證誤差 $\mathbf{e}=\mathbf{b}-\mathbf{p}$ 同時正交於 $\mathbf{u}_1$ 與 $\mathbf{u}_2$

```python
u1 = sp.Matrix([1, 2, 2])
u2 = sp.Matrix([2, -1, 0])
b = sp.Matrix([1, 1, 1])

# 1. Orthogonality
print("1. u1 · u2 =")
display_large(u1.dot(u2))
print("Orthogonal?", u1.dot(u2) == 0)

# 2. Normalize
q1 = sp.simplify(u1 / u1.norm())
q2 = sp.simplify(u2 / u2.norm())
print("\n2. Orthonormal q1, q2:")
display_large(q1)
display_large(q2)
print("q1 · q2 =")
display_large(sp.simplify(q1.dot(q2)))
print("||q1||, ||q2|| =")
display_large(sp.Matrix([q1.norm(), q2.norm()]))

# 3. Projection onto span{u1, u2}
A = sp.Matrix.hstack(u1, u2)
xhat = (A.T * A).inv() * A.T * b
p = sp.simplify(A * xhat)
print("\n3. Projection p =")
display_huge(p)

# Equivalent via orthonormal coeffs
p_alt = sp.simplify(q1.dot(b) * q1 + q2.dot(b) * q2)
print("Same via orthonormal expansion?")
display_huge(p_alt)
print("Match?", p == p_alt)

# 4. Error orthogonal to both
e = sp.simplify(b - p)
print("\n4. Error e =")
display_huge(e)
print("e · u1 =")
display_large(sp.simplify(e.dot(u1)))
print("e · u2 =")
display_large(sp.simplify(e.dot(u2)))
```

## 小結

| 概念 | SymPy / 公式 | 重點 |
|------|--------------|------|
| 內積 | `u.dot(v)` 或 `u.T*v` | $\mathbf{u}\cdot\mathbf{v}=0$ $\Leftrightarrow$ 正交 |
| 長度 | `u.norm()` | $\|\mathbf{u}\|=\sqrt{\mathbf{u}\cdot\mathbf{u}}$ |
| Gram 矩陣 | `A.T*A` | 對角＝長度平方；非對角＝內積 |
| 投影到直線 | $(\mathbf{a}\cdot\mathbf{b}/\|\mathbf{a}\|^2)\mathbf{a}$ | 誤差 $\perp\mathbf{a}$ |
| 投影到 $\operatorname{Col}(A)$ | $A(A^{\mathsf{T}}A)^{-1}A^{\mathsf{T}}\mathbf{b}$ | $A^{\mathsf{T}}\mathbf{e}=\mathbf{0}$ |
| 正交矩陣 | $Q^{\mathsf{T}}Q=I$ | $Q^{-1}=Q^{\mathsf{T}}$，保長 |
| Gram–Schmidt / QR | 手算或 `A.QRdecomposition()` | 任意獨立組 → 正交／正規正交 |

**記住：**

- 正交 = 內積為零 = 幾何上垂直
- 非零正交集合 $\Rightarrow$ 線性獨立（但獨立不一定正交）
- 正規正交基底讓展開係數變成單純的內積
- $\operatorname{Row}(A)\perp N(A)$，$\operatorname{Col}(A)\perp N(A^{\mathsf{T}})$
