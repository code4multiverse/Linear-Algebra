## 題組八：arctan 與 atan2 的分別（第 51–55 題）

下列每一題請分別說明若僅用 arctan ,$(\tan^{-1})$ 計算會出現的象限錯誤，以及用 atan2 計算可得正確結果的原因，並直接寫出正確的答案：

**51. 計算點 $(-1, 1)$（第二象限）與原點的夾角。**  
arctan($1/(-1)$) = arctan($-1$) = $-\dfrac{\pi}{4}$，僅用 arctan 會得到第四象限的答案；  
atan2(1, -1) = $\dfrac{3\pi}{4}$，正確，因為 atan2 考慮了 $x$、$y$ 的正負。  
**答：$\dfrac{3\pi}{4}$**


**52. 計算點 $(-2, -3)$（第三象限）與原點的夾角。**  
arctan($\dfrac{-3}{-2}$) = arctan($1.5$) ≈ $0.9828$，僅用 arctan 會給出第一象限答案；  
atan2($-3$, $-2$) ≈ $-2.1588$ 或 $2.1588-\pi$。  
**答：約 $-2.16$ 或 $2.16$ 弧度**


**53. 計算點 $(0, -5)$（第四象限）與原點的夾角。**  
arctan($\dfrac{-5}{0}$)：無法計算（未定義）；  
atan2($-5$, $0$) = $-\dfrac{\pi}{2}$  
**答：$-\dfrac{\pi}{2}$**


**54. 計算點 $(4, -4)$（第四象限）與原點的夾角。**  
arctan($-4/4$) = arctan($-1$) = $-\dfrac{\pi}{4}$，僅用 arctan 得到第四象限正確，但若 $x$ 是負則錯；  
atan2($-4$, $4$) = $-\dfrac{\pi}{4}$，正確。  
**答：$-\dfrac{\pi}{4}$**


**55. 計算點 $(-5, 0)$（在 x 軸負半軸）與原點的夾角。**  
arctan($0/(-5)$) = arctan($0$) = $0$，僅用 arctan 會得到第一象限答案；  
atan2($0$, $-5$) = $\pi$，正確。  
**答：$\pi$**
