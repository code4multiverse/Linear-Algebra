```python
#46

x, y = sp.symbols('x y', real=True)

# explain why equation |z - 1| = 2 to be
expr = sp.Eq(sp.sqrt((x-1)**2 + y**2), 2)

display(expr)

# To plot using sympy.plot_implicit:
sp.plot_implicit(expr, (x, -3, 5), (y, -3, 3), 
                 xlabel="Re(z)", ylabel="Im(z)", 
                 title=r"Plot of $|z-1|=2$", 
                 line_color='blue', show=True, aspect_ratio=(1,1))
```

```python
#49

# Draw the locus where |z| = |z - 2|, i.e., the set of points in the complex plane where the distance to 0 equals the distance to 2.
# This is the perpendicular bisector of the segment joining 0 and 2, i.e., Re(z) = 1.

expr_eq = sp.Eq(sp.sqrt(x**2 + y**2), sp.sqrt((x - 2)**2 + y**2))

display(expr_eq)

sp.plot_implicit(expr_eq, (x, -3, 5), (y, -3, 3),
                 xlabel="Re(z)", ylabel="Im(z)",
                 title=r"Plot of $|z| = |z-2|$",
                 line_color='purple', show=True, aspect_ratio=(1, 1))
```

```python
#50

#Step 1
x, y = sp.symbols('x y', positive=True)

# Step 2: Define the original equation
eq = sp.Eq(sp.arg(x + y * sp.I), sp.pi/4)
print("Original Equation:")
display(eq)

# Step 3: Rewrite the argument using tan/inverse trig functions
# SymPy will automatically simplify arg(x + iy) to atan2(y, x)
eq_rewritten = eq.rewrite(sp.atan2)
display(eq_rewritten)

# Step 4: Apply tangent to both sides to clear the atan2 function
# sp.tan(sp.atan2(y, x)) simplifies directly to y/x when x > 0
eq_tan = sp.Eq(sp.tan(eq_rewritten.lhs), sp.tan(eq_rewritten.rhs))
display(eq_tan)

# Step 5: Solve for y to get the final equation form
final_eq = sp.solve(eq_tan, y, dict=True)[0]
output_eq = sp.Eq(y, final_eq[y])

print("\nResulting Equation:")
display(output_eq)
```
